```python
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
data1=datasets.make_gaussian_quantiles(mean=[5,5], cov=0.1, n_samples=1000, 
                                      n_features=2, n_classes=3, 
                                      shuffle=True, random_state=None)

data2=datasets.make_gaussian_quantiles(mean=[1,2], cov=0.1, n_samples=1000, 
                                      n_features=2, n_classes=3, 
                                      shuffle=True, random_state=None)


data3=datasets.make_gaussian_quantiles(mean=[4,3], cov=0.01, n_samples=100, 
                                      n_features=2, n_classes=3, 
                                      shuffle=True, random_state=None)


ax=plt.axes()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)



size=8
plt.scatter(data1[0][:,0],data1[0][:,1],s=size,alpha=0.5,c='blue')
plt.scatter(data2[0][:,0],data2[0][:,1],s=size,alpha=0.5,c='green')
plt.scatter(data3[0][:,0],data3[0][:,1],s=size,alpha=0.5,c='red')


ax.text(0.5, 2.9, "celltype1", va='center')
ax.text(3.5, 5.5, "celltype2", va='center')
ax.text(4, 2.5, "senescent cell", va='center')



plt.xticks([])
plt.yticks([])
# plt.title("cell types")
# plt.scatter(data3[0][:,0],data3[0][:,1],c='red',s=size,alpha=0.5)
# plt.scatter(data4[0][:,0],data4[0][:,1],c='red',s=size,alpha=0.5)
```

```python
data4=datasets.make_gaussian_quantiles(mean=[5,4.5], cov=0.01, n_samples=50, 
                                      n_features=2, n_classes=3, 
                                      shuffle=True, random_state=None)

data5=datasets.make_gaussian_quantiles(mean=[1.6,1.5], cov=0.01, n_samples=50, 
                                      n_features=2, n_classes=3, 
                                      shuffle=True, random_state=None)


ax=plt.axes()
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

size=8
plt.scatter(data1[0][:,0],data1[0][:,1],s=size,alpha=0.5,c='blue')
plt.scatter(data2[0][:,0],data2[0][:,1],s=size,alpha=0.5,c='green')


ax.text(0.5, 2.9, "celltype1", va='center')
ax.text(3.3, 5.5, "celltype2", va='center')
# ax.text(3, 3, "senescent cell", va='center')

plt.scatter(data4[0][:,0],data4[0][:,1],c='red',s=size,alpha=0.5,label='senescent cell')
plt.scatter(data5[0][:,0],data5[0][:,1],c='red',s=size,alpha=0.5)

plt.legend()
plt.xticks([])
plt.yticks([])
# plt.arrow(3.6,2.7,-1.4,-0.85,width=0.05,head_width=0.2,color='red')
# plt.arrow(3.8,3.2,0.8,0.85,width=0.05,head_width=0.2,color='red')
```
